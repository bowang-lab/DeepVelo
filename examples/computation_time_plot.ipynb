```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/h/haotian/Code/DeepVelo/data/ComputationTime.csv")
data
```

```python
# bar plot
# labels = data["computation time"].tolist()
scvelo_cpu = data["scVelo (cpu)"].tolist()
scvelo_8cpu = data["scVelo (8 cpu)"].tolist()
deepvelo_cpu = data["DeepVelo Complete (cpu)"].tolist()
deepvelo_gpu = data["DeepVelo Complete (cpu + gpu)"].tolist()
labels = [
    "Hindbrain (13501 cells)",
    "Pancreatic (3696 cells)",
    "Dentate Gyrus (2930 cells)",
]
```

```python


x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 5), dpi=300)
rects1 = ax.bar(
    x - width / 2 - width,
    scvelo_cpu,
    width,
    color="#B6D7E4",
    label="scVelo (cpu)",
)
rects2 = ax.bar(
    x - width / 2,
    scvelo_8cpu,
    width,
    color="#808080",
    label="scVelo (8 cpu)",
)
rects3 = ax.bar(
    x + width / 2,
    deepvelo_cpu,
    width,
    color="#4A69D9",
    label="DeepVelo (cpu)",
)
rects4 = ax.bar(
    x + width / 2 + width,
    deepvelo_gpu,
    width,
    color="#75167C",
    label="DeepVelo (cpu + gpu)",
)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Seconds")
ax.set_title("Computation Time")
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# grid
ax.grid(axis="y", linestyle="--", linewidth=0.5)

fig.tight_layout()

plt.show()
```
